```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
%matplotlib inline
```

```python
dataset = pd.read_csv("citibike.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
np.unique(dataset['end station id']).shape
```

```python
id_to_name = dict()
indx_to_id = dict()
id_to_indx = dict()
i = 0
for entry in dataset.iloc[:,[7,8]].values:
    if entry[0] not in id_to_name:
        id_to_name[entry[0]] = entry[1]
        indx_to_id[i] = entry[0]
        id_to_indx[entry[0]] = i
        i = i + 1
    else:
        pass
```

```python
joblib.dump(id_to_name,"./Transformed Data/id_to_name.pkl")
joblib.dump(indx_to_id,"./Transformed Data/indx_to_id.pkl")
joblib.dump(id_to_indx,"./Transformed Data/id_to_indx.pkl")
```

```python
orig_x = dataset.iloc[:,5].values
orig_y = dataset.iloc[:,6].values
lat = dataset.iloc[:,9].values}
long = dataset.iloc[:,10].values
min_lat = min(lat)
max_lat = max(lat)
min_long = min(long)
max_long = max(long)
```

```python
MAP = plt.imread("map.png")
asp_ratio = MAP.shape[0]/MAP.shape[1]
size = 8
plt.figure(figsize=(size,asp_ratio*size))
plt.title("Hiring Destination Trend")
plt.plot([orig_x,lat],[orig_y,long],'--', c = 'r',alpha = 0.1)
plt.imshow(MAP,extent=[min_lat,max_lat,min_long,max_long], aspect="auto")
plt.xlim(min_lat,max_lat)
plt.ylim(min_long,max_long)
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.plot()
```

```python
destinations = []
for ID in dataset['end station id']:
    destinations.append(id_to_indx[ID])
destinations = np.array(destinations)
```

```python
plt.hist(destinations)
```

```python
plt.bar(pd.Series(destinations).value_counts().index,pd.Series(destinations).value_counts().values)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
X = (dates.drop(columns = ['MM','YYYY','mm']).values)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,destinations, test_size = 0.33, random_state = 32)
```

```python
knn.fit(X_train,y_train)

knn.score(X_train,y_train)
```

```python
counts = pd.Series(destinations).value_counts()
```

```python
plt.pie(counts.values, labels=counts.values)
plt.show()
```

```python
mask_it = dict()
vals = counts.values
idxes = counts.index
for i in range(len(counts)):
    if vals[i] < 57 :
        mask_it[idxes[i]] = True
    else:
        mask_it[idxes[i]] = False
```

```python
destinations
```

```python
new_dest = []
for i in destinations:
    if mask_it[i] == True:
        new_dest.append(-1)
    else:
        new_dest.append(i)
new_dest = np.array(new_dest)
```

```python
np.unique(new_dest)
```

```python
plt.pie(pd.Series(new_dest).value_counts().values, labels=pd.Series(new_dest).value_counts().values)
plt.show()
```
