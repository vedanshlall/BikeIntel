```python
X_train = X[:slice_idx]
y_train = Y[:slice_idx]
X_test = X[slice_idx:]
y_test = Y[slice_idx:]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report , confusion_matrix
```

```python
def show_performance(model, X_train = X_train, X_test = X_test, y_train = y_train, y_test = y_test):
    '''fit'''
    model.fit(X_train, y_train)
    
    '''Predictions'''
    
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)
    
    preds = model.predict(X)
    
    '''Metrics'''
    train_acc = model.score(X_train, y_train)
    test_acc = model.score(X_test, y_test)
    
    train_clf_report = classification_report(y_train, train_preds)
    test_clf_report = classification_report(y_test, test_preds)
    
    print("Train Data evaluation report:\n",train_clf_report,
        "\n\nTest Data evaluation report:\n", test_clf_report )
    
    train_conf = confusion_matrix(y_train, train_preds)
    test_conf = confusion_matrix(y_test, test_preds)
    
    print("Confusion Matrices: ")
    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    sns.heatmap(train_conf, annot = True)
    plt.xlabel("Train")
    plt.subplot(1,2,2)
    sns.heatmap(test_conf, annot = True)
    plt.xlabel("Test")
    plt.show()
    
    plt.figure(figsize=(20,3))
    plt.plot(range(0,186),y_train, label = "Train Values")
    plt.plot(range(186,248),y_test, label = "Test Values")
    plt.plot(range(0,186),train_preds, '--', label = "Predicted Train Values")
    plt.plot(range(186,248),test_preds, '--', label = "Predicted Test Values")
    plt.xticks(())
    plt.yticks(range(2))
    plt.xlabel("Date Line --->")
    plt.ylabel("Demand")
    plt.legend()
    plt.show()
    
    print("Failure/Fault Spectrum Analysis: ")
    ap=10
    faults = np.square(model.predict(X)-Y)
    plt.figure(figsize=(20,3))
    plt.subplot(3,1,1)
    plt.imshow(Y.reshape(1,-1), aspect=ap, cmap='gray')
    plt.xticks(())
    plt.yticks(())
    plt.ylabel("True")
    
    plt.subplot(3,1,2)
    plt.imshow(faults.reshape(1,-1), aspect=ap, cmap='gray')
    plt.xticks(())
    plt.yticks(())
    plt.ylabel("Fault")
    
    plt.subplot(3,1,3)
    plt.imshow(preds.reshape(1,-1), aspect=ap, cmap='gray')
    plt.xticks(())
    plt.yticks(())
    plt.ylabel("Prediction")
    plt.show()
    
    return model, [train_acc, test_acc]


def compare_models(model_arr):
    print("||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||")
    accs = []
    i = 0
    for model in model_arr:
        print("================================================\nIndex: ", i, "\nModel Used:\n",model)
        model, acc = show_performance(model)
        accs.append(acc)
        i = i + 1
    accs = np.array(accs)
    
    plt.figure(figsize=(10,2))
    plt.title("Accuracies Comparison")
    plt.plot(accs[:,0], label = "Train Accuracy")
    plt.plot(accs[:,1], label = "Test Accuracy")
    plt.xticks(range(len(model_arr)))
    plt.xlabel("Model Array Index")
    plt.ylabel("Accuracy Score")
    plt.legend()
    plt.show()
    print("||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||")
```

```python
knns = [KNeighborsClassifier(n_neighbors=i) for i in range(1,5)]
rfcs = [RandomForestClassifier(n_estimators=i) for i in [10,50,100,500,1000]]
lrs = [LogisticRegression()]
```

```python
for model_array in [knns,rfcs,lrs]:
    compare_models(model_array)
```
