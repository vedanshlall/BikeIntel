```python
for station in Stations:
    path = dpath.data_avenue+dpath.stationswise+station
    station_data = pd.read_csv(path+"/"+station+".csv")
    hrs = station_data['hh'].values
    hrs_1 = []
    
    for hr in hrs:

        if hr>=0 and hr<3:
            hr = '0-3'

        elif hr>=3 and hr<6:
            hr = '3-6'

        elif hr>=6 and hr<9:
            hr = '6-9'

        elif hr>=9 and hr<12:
            hr = '9-12'

        elif hr>=12 and hr<15:
            hr = '12-15'

        elif hr>=15 and hr<18:
            hr = '15-18'

        elif hr>=18 and hr<21:
            hr = '18-21'

        elif hr>=21 and hr<24:
            hr = '21-24'
    
        hrs_1.append(hr)
        
    hrs = pd.Series(hrs_1, name="hh")
    station_data['Day'] = station_data['Day'].astype(object)
    
    pd.concat([station_data.iloc[:,:3],hrs, station_data['Day']], axis = 1).to_csv(path+"/"+station+"_hr_binned.csv", index = False)
    
    print(station+"_hr_binned.csv processed successfully...")
```

```python
for station in dpath.stations:
    print("Processing "+station+"_hr_binned.csv ...", end = "  ")
    ndata= []
    path = dpath.data_avenue+dpath.stationswise+station+"/"
    data = pd.read_csv(path+station+"_hr_binned.csv")
    yrs = data['YY'].values
    for yy in range(yrs.min(), yrs.max()+1):
        data_yy = data[data['YY'] == yy]
        for mm in range(1,13):
            data_yy_mm = data_yy[data_yy['MM'] == mm]
            
            if data_yy_mm.shape[0] == 0:
                continue
            
            dd_max = data_yy_mm['DD'].values.max()
            
            for dd in range(1,dd_max+1):
                data_yy_mm_dd = data_yy_mm[data_yy_mm['DD'] == dd]  
                if data_yy_mm_dd.shape[0] == 0:
                    continue
                    
                day = str(np.unique(data_yy_mm_dd['Day'].values)[0])
                for hr in ['0-3', '3-6', '6-9', '9-12', '12-15', '15-18', '18-21', '21-24']:
                    data_yy_mm_dd_hh = data_yy_mm_dd[data_yy_mm_dd['hh'] == hr]
                    bookings = data_yy_mm_dd_hh.shape[0]
                    row = [yy, mm, dd, day, hr, bookings]
                    ndata.append(row)
    ndata = np.array(ndata)
    ndata = pd.DataFrame(ndata, columns = ["YY", "MM", "DD", "Day", "HH", "Bookings"])
    ndata.to_csv(path+station+"_bookings.csv", index = False)
    print("Done!")
```
